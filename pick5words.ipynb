```python
import pytagcloud
from konlpy.tag import Okt
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import collections
from nltk import bigrams, word_tokenize
from nltk.util import ngrams
from tqdm import tqdm
from nltk import ConditionalFreqDist
from nltk.probability import ConditionalProbDist, MLEProbDist
import warnings
warnings.simplefilter("ignore")

tagger = Okt()

def tokenize(data):
    tokens = ['/'.join(t) for t in tagger.pos(data)]
    return tokens
```

```python
with open('totalWordFreq.json', 'r') as jsonFile:
    totalWordFreq = json.load(jsonFile)
```

```python
df = pd.read_csv('27youtubers.csv', encoding = 'utf-8-sig')
```

```python
youtuber = [name for name in df.youtuber]
```

```python
finalWordFreq = {}
for name in youtuber:
    finalWordFreq[name] = totalWordFreq[name]
```

```python
def wordFreq(youtuber):
    with open('comments_json\{0}.json'.format(youtuber), 'r') as jsonFile:
        comment = json.load(jsonFile)
    tokens = tokenize(comment)
    print("Successfully Get Tokens of {0}".format(youtuber))
    tokens = [token for token in tokens if '/Noun' in token or '/Adjective' in token]
    tokenCount = collections.Counter(tokens)
    sortedToken = sorted(tokenCount.items(), key=lambda x: x[1], reverse=True)
    sortedToken = sortedToken[:100] ########상위 몇 개 개수 논의
    sortedTokenDict = {}
    for i in range(len(sortedToken)):
        sortedTokenDict[sortedToken[i][0]] = sortedToken[i][1]
        
    return sortedTokenDict
```
